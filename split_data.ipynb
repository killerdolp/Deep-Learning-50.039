```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load dataset
df = pd.read_csv("creditcard.csv")

# Preserve class ratio when possible for test and validation splits
stratify_col = df["Class"] if "Class" in df.columns else None

# 80% train, 20% temporary split
train_df, temp_df = train_test_split(
    df,
    test_size=0.20,
    random_state=42,
    stratify=stratify_col,
)

# Split remaining 20% equally into 10% test and 10% validation
temp_stratify = temp_df["Class"] if "Class" in temp_df.columns else None
test_df, valid_df = train_test_split(
    temp_df,
    test_size=0.50,
    random_state=42,
    stratify=temp_stratify,
)

# Save split files
train_df.to_csv("train_creditcard.csv", index=False)
test_df.to_csv("test_creditcard.csv", index=False)
valid_df.to_csv("valid_creditcard.csv", index=False)

print("Saved:")
print(f"train_creditcard.csv: {len(train_df)} rows")
print(f"test_creditcard.csv: {len(test_df)} rows")
print(f"valid_creditcard.csv: {len(valid_df)} rows")
```
